## Função que separa o dataframe em dataframes com menos atributos

```python
def separaDimensoes(dataFrame, dimensoes):
    """
    Função que separa as dimensões de um DataFrame em DataFrames separados.

    Parâmetros:
    df (pd.DataFrame): DataFrame com as dimensões e seus dados.
    dimensoes (dict): Dicionário com as dimensões já separadas.

    Retorna:
    dfs (dict): Dicionário com os DataFrames separados.
    """

    # Inicializa o dicionário de dimensões
    dfs = {}
    
    for dim in dimensoes:
        # Cria um DataFrame com as colunas da dimensão
        dfs[dim] = dataFrame[dimensoes[dim]]
    
    return dfs
```

```python
dfs = separaDimensoes(df, dimensoes)
```

```python
def separaDimensoes(dataFrame, dimensoes):

    # Inicializa o dicionário de dimensões
    dfs = {}
    
    for dim, dados in dimensoes.items():
        columns = dados['columns'].keys()
        
        # Cria um DataFrame com as colunas da dimensão
        df_dim = dataFrame[list(columns)]
        dfs[dados['tableNameInDataBase']] = df_dim
    
    return dfs
```
