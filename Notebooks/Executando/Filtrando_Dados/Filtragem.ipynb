# Filtragem dos arquivos csv
Essa filtragem é necessária para que seja trabalhado apenas com os atributos de interesse. Também poderá ser feita a remoção de algumas linhas para que não seja necessário um alto consumo de processamento.

## Variáveis de interesse

```python


var_interesse_ordenada = [
    'ANO_CMPT', 'MES_CMPT', 'DT_INTER', 'DT_SAIDA', 'CEP', 'MUNIC_RES', 'MUNIC_MOV', 
    'CGC_HOSP', 'CNES', 'NASC', 'SEXO', 'IDADE', 'COD_IDADE', 'NACIONAL', 'INSTRU', 
    'RACA_COR', 'ETNIA', 'CBOR', 'MORTE', 'UTI_MES_TO', 'MARCA_UTI', 'VAL_UTI', 
    'PROC_SOLIC', 'PROC_REA', 'VAL_SH', 'VAL_SP', 'N_AIH', 'VAL_TOT', 'INFEHOSP', 
    'IND_VDRL', 'DIAG_PRINC', 'DIAG_SECUN', 'DIAGSEC1', 'DIAGSEC2', 'DIAGSEC3', 
    'DIAGSEC4', 'DIAGSEC5', 'DIAGSEC6', 'DIAGSEC7', 'DIAGSEC8', 'DIAGSEC9', 'CID_MORTE'
]
```

## Filtragem

```python
nomes = [
    'RDSP1402',
    'RDSP1502',
    'RDSP1602',
    'RDSP1702',
    'RDSP1802',
    'RDSP1902',
    'RDSP2002',
    'RDSP2102',
    'RDSP2202',
    'RDSP2302',
    'RDSP2402'
]

len(nomes)
```

```python
import pandas as pd
```

```python
caminho_base = '/home/ferdanielinux/desktop-linux/scripts-python/proj-aries/Dados/csv/fevereiro/'
```

```python
data_frames = []
```

### Inserindo todos os dfs em uma lista

```python
for nome in nomes:
    caminho_csv = f'{caminho_base}{nome}.csv'
    df = pd.read_csv(caminho_csv, sep=';', encoding='latin-1', usecols=var_interesse_ordenada, nrows=100)
    data_frames.append(df)
    
len(data_frames)
```

## Concatenando e transformando o df em csv

```python
concat_dfs = pd.concat(data_frames, ignore_index=True)
```

```python
resumo_fev = concat_dfs.to_csv('/home/ferdanielinux/desktop-linux/scripts-python/proj-aries/Dados/csv/fevereiro/concat_fev_ord.csv', sep=';', index=False, encoding='latin-1')
```
