# Conectando um um banco remoto ao Python

```python
import pandas as pd
from sqlalchemy import create_engine
```

## Conectando com o banco de dados

```python
# Configurações da conexão
usuario = 'postgres'
senha = '123,258D'
host = 'localhost'
porta = '5432'
banco_de_dados = 'my_test'

# URL de conexão
url = f"postgresql://{usuario}:{senha}@{host}:{porta}/{banco_de_dados}"
```

```python
engine = create_engine(url)

# Teste a conexão
try:
    connection = engine.connect()
    print("Conexão bem-sucedida!")
except Exception as e:
    print(f"Erro ao conectar: {e}")
finally:
    connection.close()
```

## Inserindo arquivo parquet no banco de dados conectado

```python
arquivo = '/home/ferdanielinux/desktop-linux/scripts-python/proj-aires/Dados/SIH/stage/parquet/2018/RDSP1802.parquet'
```

```python
df = pd.read_parquet(arquivo)
```

## Escrevendo o dataframe no banco de dados conectado

```python
df.head(20).to_sql('teste2', engine, if_exists='replace', index=True)
```
