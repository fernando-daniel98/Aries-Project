# Copiando um df para um banco de dados já criado

## Criando um df pronto para inserir ao banco de dados

### Bibliotecas utilizadas

```python
import pandas as pd
import csv
```

### Funções utilizadas

```python
# Função que pulará algumas linhas iniciais se preciso (Posso usar para pular o cabeçalho) 
def pular_n_linhas(leitura, n):
  for i in range(n):
    next(leitura)
      

# Função que retorna o último índice útil do dataframe
def num_lin(nome_arquivo, formato, n):

    indice = 0

    try:
      with open(nome_arquivo, 'r', encoding = formato) as arq:
        leitor_csv = csv.reader(arq)
        pular_n_linhas(leitor_csv, n)

        for linha in leitor_csv:
          if linha[0].startswith('Total'):
            break
          else:
            indice += 1

        return indice

    except FileNotFoundError:
        print(f"O arquivo '{nome_arquivo}' não foi encontrado.")
    except Exception as e:
        print(f"Ocorreu um erro: {e}")
        return None
```

```python
arquivo = 'C:/Users/Fernando/Desktop/Compromissos/IC - Banco de Dados/Dados/sih_cnv_nibr101851201_75_169_32.csv'

formato = 'ISO-8859-1'

# Notei um padrão que todo arquivo .csv baixado possue as 3 primeiras linhas referentes à consulta
num_linhas_puladas = 3;

qtd_linhas_uteis = num_lin(arquivo, formato, num_linhas_puladas)

df = pd.read_csv(arquivo, sep = ';', encoding = 'ISO-8859-1', skiprows = 3, nrows = qtd_linhas_uteis);
```
